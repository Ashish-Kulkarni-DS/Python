```python
import seaborn as sns
import numpy as np 
import pandas as pd
```

```python
iris = pd.read_csv("F:\Data Science\Datasets\Iris\IRIS.csv")
iris
```

## Lineplot

It represents the change in a quantity with respect to another quantity. 

This variation is usually plotted in a two-dimensional XY plane. If the relation including any two measures
can be expressed utilizing a straight line in a graph, then such graphs are called linear
graphs. 

Thus, the line plot is also called a linear plot.

```python
print(list(iris)) 
```

```python
iris.columns
```

```python
sns.lineplot(x = 'sepal_length', y = 'sepal_width', data = iris)
```

```python
sns.lineplot(x = 'sepal_length', y = 'sepal_width', data = iris, hue = 'species')
```

## Barplot
A barplot is basically used to aggregate the categorical data according to some methods and by default it’s the mean. 

It can also be understood as a visualization of the group by action. To use this plot we choose a categorical column for the x-axis and a numerical column for the y-axis, and we see that it creates a plot taking a mean per categorical
column.

```python
sns.barplot(x = 'species', y = 'sepal_length', data = iris)
```

## Countplot

A countplot method is used to show the counts of observations in each categorical bin
using bars.

```python
sns.countplot(x = 'species', data = iris)
```

## Boxplot

Box Plot is the visual representation of the depicting groups of numerical data through their quartiles. Boxplot is also used for detect the outlier in data set. 

It captures the summary of the data efficiently with a simple box and whiskers and allows us to compareeasily across groups. Boxplot summarizes a sample data using 25th, 50th and 75th
percentiles. 

These percentiles are also known as the lower quartile, median and upper quartile.

```python
sns.boxplot(x = 'species', y = 'sepal_length', data = iris)
```

```python
sns.boxplot(x = 'species', y = 'sepal_width', data = iris)
```

## Violinplot

A violin plot plays a similar activity that is pursued through whisker or box plot do. As it shows several quantitative data across one or more categorical variables.

It can be an effective and attractive way to show multiple data at several units. A “wide-form” DataFrame helps to maintain each numeric column which can be plotted on the graph.


```python
sns.violinplot(x = 'species', y = 'sepal_length', data = iris)
```

```python
sns.violinplot(x = 'species', y = 'sepal_width', data = iris)
```

## Histplot (Histogram)

```python
sns.histplot(x = 'sepal_width', data = iris)
```

```python
sns.histplot(x = 'sepal_width', hue = 'species', data = iris)
```

## Distplot

It is used basically for univariant set of observations and visualizes it through a histogram
i.e. only one observation and hence we choose one particular column of the dataset.

```python
sns.distplot(iris['sepal_width'])
```

```python
sns.distplot(iris.sepal_width)
```

## Pairplot

To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function. This shows the relationship for (n, 2) combination of variable in a DataFrame as a matrix of plots and the diagonal plots are the univariate plots.


```python
sns.pairplot(data = iris, hue = 'species')
```

## KDEplot

KDE Plot described as Kernel Density Estimate is used for visualizing the Probability Density of a continuous variable. It depicts the probability density at different values in a continuous variable. 

We can also plot a single graph for multiple samples which helps in more efficient data visualization.


```python
sns.kdeplot(x = iris.sepal_width, data = iris)
```

```python
sns.kdeplot(y = iris.sepal_width, data = iris)
```

```python
sns.kdeplot(data = iris)
```

```python
sns.kdeplot(x = iris.sepal_width, data = iris, hue = 'species')
```

```python
sns.kdeplot(x = iris.sepal_width, data = iris, hue = 'species', fill = 'stack')
```

```python
sns.kdeplot(x = iris.sepal_width, data = iris, hue = 'species', fill = True)
```

```python
sns.kdeplot(x = iris.sepal_width, data = iris, hue = 'species', multiple = 'fill')
```
